```python
display(dbutils.fs.ls("/mnt/adlsgen2"))
df = spark.read.csv("/mnt/adlsgen2/customers_data.csv", header=True, inferSchema=True)
df.show(5)
from pyspark.sql.types import StructType, StructField, StringType, IntegerType

display(dbutils.fs.ls("/mnt/adlsgen2"))

df = spark.read.csv(
    "/mnt/adlsgen2/customers_data.csv",
    header=True,
    inferSchema=True
)
df.show(5)

# Define the schema for the JSON file
json_schema = StructType([
    StructField("customer_id", IntegerType(), True),
    StructField("customer_name", StringType(), True),
    StructField("customer_email", StringType(), True)
])

df_json = spark.read.schema(json_schema).json("/mnt/adlsgen2/customer_data.json")
df_json.show(5)


df_tsv = spark.read.csv("/mnt/adlsgen2/customers_data.tsv", header=True, sep="\t", inferSchema=True)
df_tsv.show(5)


df_txt = spark.read.text("/mnt/adlsgen2/customers_data.txt")
df_txt.show(5)



```

```python
df.createOrReplaceTempView("customers_csv_view")
df_json.createOrReplaceTempView("customers_json_view")
df_tsv.createOrReplaceTempView("customers_tsv_view")
df_txt.createOrReplaceTempView("customers_txt_view")
spark.sql("SELECT * FROM customers_csv_view WHERE email IS NOT NULL").show()
```

```python
spark.sql("SELECT COUNT(*) FROM customers_csv_view").show()
```

```python
spark.sql(
    "SELECT state, COUNT(*) AS total FROM customers_csv_view GROUP BY state"
).show()
```

```python
spark.sql("""
SELECT c.first_name, c.last_name, c.email, j.customer_email
FROM customers_csv_view c
JOIN customers_json_view j
ON c.customer_id = j.customer_id
""").show()
```

```python
spark.sql("""
  CREATE OR REPLACE TABLE customers_delta
  USING DELTA
  AS SELECT * FROM customers_csv_view
  WHERE email IS NOT NULL
""")
spark.sql("SELECT * FROM customers_delta LIMIT 5").show()
spark.sql("SELECT COUNT(*) FROM customers_delta").show()
```

```python
df.write.format("delta").mode("overwrite").save("/mnt/adlsgen2/delta/customers")
spark.sql("""
  CREATE TABLE IF NOT EXISTS delta_customers_saved
  USING DELTA
  LOCATION '/mnt/adlsgen2/delta/customers'
""")
```

```python
spark.sql("SELECT * FROM delta_customers_saved LIMIT 5").show()
```
